```python

import numpy as np
from scipy.integrate import quad

# (a) ∫₄⁹ 3√x dx
f_a = lambda x: 3 * np.sqrt(x)
result_a, _ = quad(f_a, 4, 9)
print("a. ∫₄⁹ 3√x dx =", result_a)

# (b) ∫₁ᵉ ln(x) dx
f_b = lambda x: np.log(x)
result_b, _ = quad(f_b, 1, np.e)
print("b. ∫₁ᵉ ln(x) dx =", result_b)

# (c) ∫₀¹ arccos(x) dx
f_c = lambda x: np.arccos(x)
result_c, _ = quad(f_c, 0, 1)
print("c. ∫₀¹ arccos(x) dx =", result_c)

# (d) ∫₋¹¹ π cos(πx/2) dx
f_d = lambda x: np.pi * np.cos(np.pi * x / 2)
result_d, _ = quad(f_d, -1, 1)
print("d. ∫₋¹¹ π cos(πx/2) dx =", result_d)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 100)
y = x**3 - 5*x**2 + 30

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x^3 - 5x^2 + 30')
plt.grid(True)
plt.show()
```
